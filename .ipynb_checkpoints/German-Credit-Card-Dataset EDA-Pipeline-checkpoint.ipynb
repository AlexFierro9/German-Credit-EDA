Here are the key observations of this violin plot:

    *   "Own" housing has the vast majority of loans concentrated at very low amounts (near 0).
    *   "Free" and "rent" housing categories show distributions shifted towards higher loan amounts compared to "own", with higher median values and greater spread.

2.  **Impact of Credit Risk Depends on Housing Status:** The difference in loan amount distribution between "Good Credit" and "Bad Credit" varies depending on the housing category.
    *   **"Own" Housing:** There is very little difference in the loan amount distribution between "Good Credit" and "Bad Credit". Both groups are heavily concentrated at extremely low loan amounts.

3.  **Density Shapes Reveal Concentration Differences:**
    *   In "own", both credit groups show a sharp peak in density near 0.
    *   In "free" and "rent", the "Good Credit" group often shows a single peak at a moderate loan amount, while the "Bad Credit" group tends to have density more spread out across lower values or peaked closer to 0.

4.  **Box Plots Reinforce Median and IQR Differences:** The box plots inside the violins clearly show the median (dashed line) and interquartile range (box boundaries). This visually confirms the observations about medians being higher for "Good Credit" in "free" and "rent" categories, and medians being very low for both groups in the "own" category.

5.  **Outliers Exist Across Categories:** All categories and credit risk groups show some individual loan amounts (represented by dots) that are significantly higher than the main body of the distribution, extending up towards 20k or more. This suggests some large loans are granted regardless of housing status or credit score, although the density of such loans is low.

In conclusion, while loan amounts are generally lowest for homeowners, for those living "free" or "renting", having "Good Credit" is associated with significantly higher loan amounts compared to having "Bad Credit". Credit risk seems to play a much less differentiating role in the *amount* borrowed for those who own their homes, where most loans are very small regardless of credit score.

**Graph 1: Log Scale Distribution Plot**

*   **Visualization Elements:**
    *   **Histograms:** The shaded bars represent the histograms for each group. They show the frequency (or density) of values falling within specific bins of the logarithmic credit amount. You can see that the bins are relatively wide.
    *   **Kernel Density Estimates (KDEs):** The smooth curves are the KDEs for each group. These curves provide a smoothed estimate of the probability density function for each group. They give a better sense of the overall shape of the distribution than the raw histograms, especially with wider bins.
    *   **Rug Plots:** The small vertical lines at the bottom are rug plots. Each line represents an individual data point. This gives a raw view of the data distribution along the x-axis.


*   **Analysis:**
    *   **Overall Distribution:** Both Good Credit and Bad Credit groups show a roughly unimodal (single peak) distribution for the logarithm of credit amount.
    *   **Peak Location:** The peak of the distribution for Good Credit appears to be slightly shifted to the right compared to Bad Credit. This suggests that individuals with Good Credit tend to have *slightly higher* (on average) logarithmic credit amounts than individuals with "Bad Credit."
    *   **Shape of Distribution:** The distributions appear somewhat similar in shape, although the Good Credit distribution might be slightly wider or have a longer tail on the higher side. The KDEs help to see this more clearly.
    *   **Overlap:** There is significant overlap between the distributions of Good Credit and Bad Credit. This indicates that credit amount alone is not a perfect predictor of credit risk. Many individuals with Good Credit have credit amounts similar to those with Bad Credit, and vice versa.
    *   **High-End Credit Amounts:** There seems to be a greater proportion of individuals with Good Credit having higher logarithmic credit amounts (towards the right side of the plot) compared to those with Bad Credit. This is evident in both the histograms and the KDEs.
    *   **Low-End Credit Amounts:** Similarly, there might be a slightly greater proportion of individuals with Bad Credit having lower logarithmic credit amounts (towards the left side of the plot).

**Plot 2: Credit Amount Frequency distribuition**

*   **Groups:** The data is split into two groups: Good Credit (red) and Bad Credit (green).
*   **Visualization Elements:**
    *   **Histograms:** The shaded areas represent the histograms for each group.
    *   **Kernel Density Estimates (KDEs):** The smooth curves are the KDEs for each group.

*   **Analysis of Plot 2:**
    *   **Skewness:** This plot clearly shows that the raw credit amount data is highly right-skewed. The majority of individuals have relatively low credit amounts, and there's a long tail extending towards higher credit amounts. This is a common pattern for financial data like credit amounts.
    *   **Comparison with Plot 1:** This plot illustrates *why* the logarithm was likely taken in the first plot. The skewness in the raw data makes it difficult to see the details of the distribution at the lower credit amounts and compare the groups effectively. The logarithmic plot provides a more balanced view.
    *   **Peak Location:** The peak of both distributions is concentrated at the lower end of the credit amount scale (around 0 to 5000).
    *   **Separation:** Similar to the first plot, there is significant overlap between the two groups in the raw credit amount distribution. It's even harder to see clear differences between the groups in this raw plot due to the strong skewness.
    *   **Higher Credit Amounts:** You can see that the tails of the distributions extend to similar high credit amounts for both groups, but it's difficult to determine the proportion of individuals in each group at these higher values from this plot alone.

## 3. Data Preprocessing

One Hot Encodinng to Transform data to transform categorical data to dummy variables

```python
def one_hot_encoder(df, nan_as_category = False):
    original_columns = list(df.columns)
    categorical_columns = [col for col in df.columns if df[col].dtype == 'object']
    df = pd.get_dummies(df, columns= categorical_columns, dummy_na= nan_as_category, drop_first=True)
    new_columns = [c for c in df.columns if c not in original_columns]
    return df, new_columns
```

```python
df_credit['Saving accounts'] = df_credit['Saving accounts'].fillna('no_inf')
df_credit['Checking account'] = df_credit['Checking account'].fillna('no_inf')

#Purpose to Dummies Variable
df_credit = df_credit.merge(pd.get_dummies(df_credit.Purpose, drop_first=True, prefix='Purpose'), left_index=True, right_index=True)
#Sex feature in dummies
df_credit = df_credit.merge(pd.get_dummies(df_credit.Sex, drop_first=True, prefix='Sex'), left_index=True, right_index=True)
# Housing get dummies
df_credit = df_credit.merge(pd.get_dummies(df_credit.Housing, drop_first=True, prefix='Housing'), left_index=True, right_index=True)
# Housing get Saving Accounts
df_credit = df_credit.merge(pd.get_dummies(df_credit["Saving accounts"], drop_first=True, prefix='Savings'), left_index=True, right_index=True)
# Housing get Risk
df_credit = df_credit.merge(pd.get_dummies(df_credit.Risk, prefix='Risk'), left_index=True, right_index=True)
# Housing get Checking Account
df_credit = df_credit.merge(pd.get_dummies(df_credit["Checking account"], drop_first=True, prefix='Check'), left_index=True, right_index=True)
# Housing get Age categorical
df_credit = df_credit.merge(pd.get_dummies(df_credit["Age_cat"], drop_first=True, prefix='Age_cat'), left_index=True, right_index=True)
```

## 5 Model Evaluation
While we will calculate accuracy, our major point of evaluations will be precision, recall and reacll, here's why I think accuracy is a poor measure for imbalanced datasets such as this one:

**The Problem with Accuracy in Imbalanced Datasets**

Accuracy is calculated as:

$$ \text{Accuracy} = \frac{\text{Number of correct predictions}}{\text{Total number of predictions}} = \frac{\text{True Positives + True Negatives}}{\text{True Positives + True Negatives + False Positives + False Negatives}} $$

In an imbalanced dataset, one class (the majority class) has significantly more instances than the other (the minority class). In credit risk, the "good credit" class is the majority class, and the "bad credit" class is the minority class (ideally, you have many more good credits than bad ones).

Here's when accuracy becomes misleading:

* **It's gets Dominated by the Majority Class:** If the majority class is, for example, 95% of the data, a model can achieve 95% accuracy by simply predicting the majority class for every instance. This model would be completely useless in identifying the minority class (bad credit), but its accuracy score would look very high.

* **It Masks Poor Performance on the Minority Class:** A high accuracy score can give a false sense of confidence, while the model is actually doing a terrible job of identifying the instances you're most interested in (the bad credits). For credit risk, failing to identify a bad credit can have significant financial consequences.

* **The Minority Class is Often the Class of Interest:** In many imbalanced scenarios, the minority class represents the "event" or "positive" case that you are trying to detect (e.g., fraud, disease, customer churn, or in this case, bad credit). A model with high accuracy but low ability to detect these events is not useful for the intended purpose.

**Illustrative Example:**

Imagine a credit risk dataset with 1000 customers:
* 950 are "Good Credit" (Majority Class)
* 50 are "Bad Credit" (Minority Class)

Consider a simple model that always predicts "Good Credit":

* **True Positives (Correctly predicted Bad Credit):** 0
* **True Negatives (Correctly predicted Good Credit):** 950
* **False Positives (Incorrectly predicted Bad Credit):** 0
* **False Negatives (Incorrectly predicted Good Credit):** 50

The accuracy of this model would be:

$$ \text{Accuracy} = \frac{0 + 950}{0 + 950 + 0 + 50} = \frac{950}{1000} = 0.95 \text{ or } 95\% $$

A 95% accuracy score looks impressive on its own. However, this model completely failed to identify any of the 50 "Bad Credit" instances, which is the critical task in credit risk.

**Why Precision and Recall are Better:**

* **Precision:** Focuses on the accuracy of the positive predictions. It answers the question: "Of all the instances the model *predicted* as 'Bad Credit', how many were *actually* 'Bad Credit'?"
    $$ \text{Precision} = \frac{\text{True Positives}}{\text{True Positives + False Positives}} $$
    A high precision means fewer false alarms.

* **Recall (Sensitivity):** Focuses on the model's ability to find all the positive instances. It answers the question: "Of all the instances that were *actually* 'Bad Credit', how many did the model *correctly identify* as 'Bad Credit'?"
    $$ \text{Recall} = \frac{\text{True Positives}}{\text{True Positives + False Negatives}} $$
    A high recall means fewer missed bad credits.

In this example above, the precision would be 0/0 (undefined or often treated as 0) and the recall would be 0/50 = 0. These metrics clearly show the model's complete failure to identify the minority class, which is much more informative than the 95% accuracy.


While accuracy is a good general-purpose metric for balanced datasets, it is misleading in imbalanced scenarios because it can be artificially inflated by the majority class. For problems like credit risk where the minority class is of high importance and represents the event we want to detect, metrics like **precision, recall, F1-score (the harmonic mean of precision and recall), and AUC-ROC** provide a much more meaningful and accurate evaluation of the model's performance, particularly in identifying the instances you care about most.

## 6. Post Training Model Feature Analysis
**Why Post-Training Feature Importance Analysis is Necessary**

Even with a powerful model like XGBoost that achieves good predictive performance, simply having a black box that makes predictions is often insufficient. Post-training feature importance analysis is essential for achieving **Explainable AI (XAI)** and addressing the need for **interpretability** and **transparency** in our credit risk model. Why do I think this is necessary?

1. **Model Understanding and Trust (Interpretability):** Understanding *why* the model makes certain predictions is critical. In credit risk, lenders and regulators need to know what factors influenced a credit decision. Feature importance analysis helps deconstruct the **black box** nature of complex models like XGBoost, revealing which variables the model deemed most influential. This interpretability builds trust in the model and the decisions it supports.

2. **Regulatory Compliance:** Financial institutions are subject to regulations (like fair lending practices) that often require justification for credit decisions. Being able to demonstrate that decisions are based on relevant and non-discriminatory factors is paramount. Feature importance provides evidence to support the **fairness** and **accountability** of the model.

3. **Business Insights and Domain Knowledge Validation:** Feature importance can validate existing domain knowledge about credit risk factors (e.g., income, credit history). It can also uncover unexpected relationships or highlight factors that were previously undervalued, leading to new business insights and strategies. This contributes to **actionable insights** derived from the model.

4. **Model Improvement and Debugging:** Understanding which features are most important can guide further model development. If irrelevant features are deemed highly important, it might indicate data issues or model biases. Conversely, if expectedly important features are not contributing significantly, it might suggest areas for feature engineering or data cleaning. This facilitates **model debugging** and **optimization**.

5. **Feature Selection and Dimensionality Reduction:** Identifying less important features can inform future data collection efforts and potentially allow for dimensionality reduction, leading to simpler, faster, and potentially more robust models.

In essence, post-training feature importance analysis moves beyond just prediction accuracy to provide **local explanations** (for individual predictions) and **global explanations** (for the overall model behavior), fulfilling the core principles of Explainable AI.

**How I'm using XGBoost's Internal Features, Permutation Importance, and SHAP**
1. **XGBoost's Internal Feature Importance:**
    * **How it works:** XGBoost calculates feature importance based on how much each feature contributes to the model's predictive power during training. Different metrics are available, such as:
        * **Weight (or Frequency):** The number of times a feature is used to split a node across all trees.
        * **Gain:** The average gain (improvement in model performance) achieved by splits using the feature. This is often the most popular metric.
        * **Cover:** The average coverage (number of instances affected by splits) where the feature is used.
    * **What it tells you:** Provides a **global explanation** of which features were most frequently used or provided the most "gain" *during training*. It gives you an initial ranking of features based on their contribution within the XGBoost structure.
    * **Strengths:** Relatively easy to calculate, built-in to XGBoost.
    * **Limitations:** Can be biased towards categorical features with many levels or features used in early splits. Doesn't directly tell you the *impact* of a feature on the prediction value, only its usage or gain within the tree structure.

2. **Permutation Importance:**
    * **How it works:** This is a model-agnostic technique. After training the model, you take a trained test dataset and evaluate the model's performance (e.g., using a suitable metric like precision or recall for your imbalanced data). Then, you randomly shuffle the values of a single feature in the test data and evaluate the model's performance again. The drop in performance after shuffling indicates the importance of that feature. You repeat this for each feature.
    * **What it tells you:** Provides a **global explanation** of how much the model's predictive performance *depends* on each feature. It's a measure of how critical each feature is to the model's overall predictive power on unseen data.
    * **Strengths:** Model-agnostic, intuitive concept, measures the impact on the final prediction metric.
    * **Limitations:** Computationally more expensive than internal feature importance, can be misleading if features are highly correlated (the model might still rely on correlated features even if one is shuffled).

3. **SHAP (SHapley Additive exPlanations):**
    * **How it works:** SHAP is a powerful game theory-based approach that assigns an "importance value" to each feature for *each individual prediction*. It explains a prediction by calculating the contribution of each feature value to the difference between the prediction and the average prediction.
    * **What it tells you:** Provides both **local explanations** (why a specific individual received a particular credit score or decision) and **global explanations** (by aggregating local explanations across many individuals). SHAP values indicate the direction and magnitude of each feature's impact on the prediction. For example, a positive SHAP value for "Credit Amount" for a specific individual might indicate that a high credit amount contributed to a higher predicted risk.
    * **Strengths:** Provides clear, interpretable local explanations. Can reveal non-linear relationships and interactions between features. Provides consistent global feature importance by averaging absolute SHAP values. Statistically sound as it's based on Shapley values from cooperative game theory.
    * **Limitations:** Can be computationally intensive, especially for large datasets and complex models. Can be challenging to explain the underlying theory to non-technical stakeholders.

**Convergence of these Methods:**

Using these three methods together provides a more comprehensive and robust picture of feature importance:

* **XGBoost's internal importance** gives a quick initial sense of features important to the model's internal structure.
* **Permutation importance** validates the overall importance of features by assessing their impact on the model's performance on unseen data.
* **SHAP** provides the most granular understanding, explaining individual predictions and revealing how features influence predictions both locally and globally, which is essential for true **XAI** and **accountability** in credit risk.

By combining these approaches, you can gain confidence in your feature importance findings, ensure the model is making decisions based on sensible factors, and effectively communicate the model's reasoning to stakeholders, addressing the crucial need for **transparency** and **explainability** in your credit risk model.

* **Mid-Tier Consistent:** Features like  Purpose_car,   Job,   Purpose_radio/TV,   Housing_own,   Check_moderate,  and  Sex_male  consistently show moderate importance – they contribute to performance, but not as drastically as the top features.
* **Lower Tier Confirmed:** Most of the  Savings  categories (except  no_inf ), some  Purpose  categories, and the age categories (except potentially  Age_cat_Young ) remain at the lower end of importance. Permuting these features doesn't hurt the model's performance much.
* **Error Bars:** The black error bars indicate the variability in importance across multiple shuffles. Larger bars mean the importance estimate is less stable. Notice how the top features have wider error bars, likely due to their large impact, while the less important features have very small error bars.

**To sum it all up**

This permutation test strongly *validates* that **Credit amount, Duration, Age, and Check_no_inf are the most critical features** for the model's predictive power. It also suggests that **Savings_no_inf is more important for actual performance than its internal XGBoost ranking implied**. Features at the bottom have minimal impact on the model's ability to correctly classify credit risk. This confirms that the features we identified with SHAP and internal importance are indeed the ones the model relies on most for accurate predictions on new data.

## 7. Conclusion

The most consistently important features across all analyses are:

1.  **Credit amount:** The size of the loan is a primary risk indicator. Larger amounts consistently increase predicted risk.
2.  **Duration:** The length of the loan term is equally critical. Longer durations consistently increase predicted risk.
3.  **Age:** Applicant age is a significant factor. Younger applicants tend to be associated with higher risk, while older applicants with lower risk.
4.  **Checking Account Status (especially  No Info ):** The status of the applicant's checking account, particularly the lack of information, is a major predictor of higher risk. Conversely, having a  rich  checking account strongly signals lower risk.

These four features are the foundation of the model's risk assessment. Messing with their values significantly impacts predictions and the model's overall performance.

Other features with moderate but consistent importance include:

*   **Savings_no_inf:** While not always highest in internal rankings, the permutation test highlighted that having no savings information is important for overall model performance.
*   **Housing_own:** Owning a home is a positive signal, reducing predicted risk.
*   **Sex_male:** Being male has a slight tendency to be associated with lower risk (important to flag for potential fairness review).
*   **Job and some Loan Purpose categories (like car/radio/TV):** These provide relevant context and contribute to predictions, but are less impactful than the top tier.

Conversely, several features consistently showed minimal importance across all methods, having very little impact on the model's predictions or performance. These include most of the other **Savings categories, Housing_rent, and certain specific Loan Purpose categories (like education, repairs, vacation/others)**.

**In essence, this XGBoost model primarily relies on core financial details (loan amount, duration, checking account status), the applicant's age, and basic demographic/housing information to assess credit risk. Savings levels and many specific loan purposes are less influential.** This provides valuable **interpretability** for our model, validates key risk factors we expected, and highlights the critical importance of obtaining clear financial data, especially regarding checking accounts, for accurate credit risk assessment. This understanding allows us to better trust the model's decisions, communicate them effectively, and focus on the most impactful data points in our processes.
